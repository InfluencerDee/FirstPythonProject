```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the diamonds.csv file from data directory

read_diamonds = pd.read_csv('diamonds.csv', index_col=0)
read_diamonds
```

```python
print(read_diamonds.describe())
print(read_diamonds.info())
print(read_diamonds.head())
print(read_diamonds.duplicated())
```

```python
numeric_cols = []
categorical_cols = []

for col in read_diamonds.columns:
    if read_diamonds[col].dtype in ['int64', 'float64']:
        numeric_cols.append(col)
    else:
        categorical_cols.append(col)

numeric_cols
#categorical_cols
```

```python
for col in numeric_cols:
    plt.figure()
    plt.hist(read_diamonds[col], bins=30)  # use default colour; donâ€™t specify colours unless asked
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
plt.scatter(read_diamonds['carat'], read_diamonds['price'], alpha=0.3)
plt.xlabel('Carat')
plt.ylabel('Price')
plt.title('Price vs Carat')
plt.show()
```
